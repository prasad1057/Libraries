```python
df = pd.DataFrame({"name":['prasad','nikhil','prasad'],
                   "toy": [np.nan,np.nan,np.nan],
                   "born": [pd.NaT,pd.Timestamp("2004-03-10"), pd.NaT]
                   })
```

```python
df
```

```python
df.dropna(how='all',axis=1)
```

```python
df.drop_duplicates(subset=['name'],keep='last')    #to keep last duplicate ele
```

```python
df.drop_duplicates(subset=['name'],keep=False)    #to remove all the duplicates
```

```python
df['name'].value_counts(dropna=False)
```

```python
df['name'].value_counts(dropna=True)
```

```python
df.isnull()
```

```python
#create a datframe which consist of only integers with 3 rows and 2 columns 
#run following datframes methods on them
df.describe()
df.min()
df.max()
df.mean()
df.corr()
df.median()
df.std()
df.count()
```
