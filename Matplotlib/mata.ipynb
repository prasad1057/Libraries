```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Grpah

```python
# For plotting simple graph
x = [5,10,15,20,25]
y = [6,12,18,24,30]

plt.plot(x,y)       #isme vvalue bhi aata he (x_label  & y_label)
plt.show()          #khali aapko grpah chahihe to
```

```python
plt.plot(x,y,color='red')       #to change the color of the graph
```

```python
#For marking in the graph
x = [5,10,15,20,25]
y = [6,12,18,24,30]
plt.plot(x,y, color='red', marker='*')
plt.show()
```

```python
# TO change the marker color we used markeredgecolor function
plt.plot(x,y, color='red', marker='*', markeredgecolor='blue')
plt.show()
```

```python
# To chnage the marker size we used markersize function
plt.plot(x,y, color='red', marker='*', markeredgecolor='blue', markersize=10)
```

```python
# to change the line we used linstyle function & it allow only to -- , it does not allow --- three hyphon
#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x,y, color='red',marker='*',linestyle='--', markeredgecolor='blue',markersize=10)
```

```python
# Shortform(color, marker, linestyle)
plt.plot(x,y,'r*--')
# here the first is r is line color & * is marker & -- is linestyle
```

```python
# For heading of the graph
plt.plot(x,y,'r*--')
plt.title('Heading')
```

```python
plt.plot(x,y,'r*--')
plt.title('Heading',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
```

```python
# to give axis name xlabel and ylabel is used
plt.plot(x,y,'r*--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# for legend of the graph(Scale)
plt.plot(x,y,'b*--')
plt.legend(['Single Element'])
```

```python
# to save the graph in ur folder
plt.plot(x,y,'b*--')
plt.savefig('firstgraph.png', dpi=200)
```

**Line Grpah**

```python
mobile = pd.read_csv('mobile.csv')
mobile.head()
```

```python
plt.plot(mobile.year, mobile.india,'b*-', label='India')
plt.plot(mobile.year,mobile.china,'r*-', label='China')
plt.plot(mobile.year,mobile.japan,'g*-', label='Japan')
plt.plot(mobile.year,mobile.america,'m*-', label="America")

plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Mobile Price Year Wise')

plt.legend()
plt.savefig('mobileprice.png',dpi=300)
plt.show()
```

**Histogram**

```python
# histogram bunch of data ko show karta he
bunch = [0,20,40,60,80,100]
x=[10,20,30,40,50,60]

plt.hist(x,bins=bunch,color='black')
plt.xlabel('Bunch')
plt.ylabel('Data')
plt.legend(["OK"])
plt.show()
```

**PieChart**

```python
y = np.array([25, 30, 35, 40, 20])  
name = ["prasad", "nikhil", "durvesh", "diva", "pranav"]

plt.pie(y, labels=name)
plt.show()
```

```python
# Explode
nameexplode = [0,0,0,0,0.2]
plt.pie(y, labels=name, explode=nameexplode)
plt.show()
```

```python
#Shadow
plt.pie(y, labels=name, explode=nameexplode,shadow=True)
plt.savefig('pie.png',dpi=300)
plt.show()
```

```python
mycolor = ['b','hotpink','black','m']
plt.pie(y, labels=name, explode=nameexplode,shadow=True, colors=mycolor)
plt.legend()
plt.show()
```

```python
test = np.random.normal(200,50,400)
plt.boxplot(test)
plt.show()
```

```python
#Multiple box plot
test1 = np.random.normal(200,50,400)
test2 = np.random.normal(100,50,300)
test3 = np.random.normal(150,100,400)
test4 = np.random.normal(250,75,400)

data = [test1,test2,test3,test4]
plt.boxplot(data)
plt.show()
```
